```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
final_df = pd.read_csv('../Modelling/Information Mart/orders_products_details.csv')
```

```python

# Set seaborn style
sns.set_theme(style="whitegrid")
```

```python
# Visualization 1: Distribution of Latency Days
plt.figure(figsize=(10, 6))
sns.histplot(final_df['latency_days'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Latency Days')
plt.xlabel('Latency Days')
plt.ylabel('Count')
plt.show()
```

```python

# Visualization 2: Countplot of Late Deliveries
plt.figure(figsize=(8, 6))
sns.countplot(x='late_delivery', data=final_df, palette='Set2', hue='late_delivery', legend=False)
plt.title('Countplot of Late Deliveries')
plt.xlabel('Late Delivery')
plt.ylabel('Count')
plt.show()
```

```python


# Visualization 3: Boxplot of List Prices (EGP) by Category
plt.figure(figsize=(12, 8))
sns.boxplot(x='category_id', y='list_price_egp', data=final_df, palette='Pastel1')
plt.title('Boxplot of List Prices (EGP) by Category')
plt.xlabel('Category ID')
plt.ylabel('List Price (EGP)')
plt.show()
```

```python
#Time Series Analysis - Number of Orders Over Time
plt.figure(figsize=(12, 6))
final_df['order_date'] = pd.to_datetime(final_df['order_date'])
orders_per_day = final_df.groupby(final_df['order_date'].dt.date).size()
sns.lineplot(x=orders_per_day.index, y=orders_per_day.values, color='skyblue')
plt.title('Number of Orders Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
#Product Analysis - Distribution of List Prices for Different Products:
plt.figure(figsize=(10, 6))
sns.histplot(final_df['list_price_egp'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of List Prices for Different Products')
plt.xlabel('List Price')
plt.ylabel('Count')
plt.show()
```

```python
#Product Category Analysis - Average List Price by Category:
plt.figure(figsize=(10, 6))
sns.barplot(x='category_id', y='list_price_egp', data=final_df, palette='Set2', hue='category_id', errorbar=None)
plt.title('Average List Price by Category')
plt.xlabel('Category ID')
plt.ylabel('Average List Price')
plt.show()

```
