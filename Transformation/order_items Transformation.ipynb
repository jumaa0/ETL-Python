```python
import pandas as pd
import os
```

```python

# Load exchange.csv and order_items.csv
exchange_df = pd.read_csv('../Staging_1/Exchange/exchange.csv')
order_items_df = pd.read_csv('../Staging_1/order_items/order_items.csv')
order_items_df
```

```python

# Merge exchange_df and order_items_df on a common column, assuming it's 'common_column'
merged_df = pd.merge(order_items_df, exchange_df[['rates.EGP']], how='cross')
merged_df
```

```python

# Calculate the new column 'list_price_egp'
merged_df['list_price_egp'] = merged_df['list_price'] * merged_df['rates.EGP']
merged_df

# Now merged_df contains the result with the new column 'list_price_egp'
```

```python
file_path = '../Staging_2/Order_items/order_items.csv'

os.makedirs(os.path.dirname(file_path), exist_ok=True)
merged_df.to_csv(file_path, index = False) 
```
